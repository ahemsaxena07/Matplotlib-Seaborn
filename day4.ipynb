```python
import numpy as np
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
s = pd.Series(np.random.standard_normal(10).cumsum(), index=np.arange(0,100,10))
s.plot()
```

```python


s = pd.Series([3, 7, 2, 9, 4], index=['a','b','c','d','e'])

s.plot(kind='bar', title='Values', color='steelblue',
       rot=0, figsize=(6,4), grid=True)
```

```python
df = pd.DataFrame(np.random.standard_normal((10,4)).cumsum(0), columns=["A","B","C","D"], index=np.arange(0,100,10))
plt.style.use('grayscale')
df.plot(kind='scatter', x="A",y="B")
```
